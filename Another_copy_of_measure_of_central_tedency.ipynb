```python
import numpy as np
from statistics import mode
```

```python
#Lists of Fahrenheit temps
week_1 = [45, 55, 49, 60, 52, 55, 58]
week_2 = [51, 49, 62, 50, 53, 56, 59]
week_3 = [60, 60, 61, 62, 57, 59, 58]
week_4 = [59, 63, 62, 61, 65, 66, 62]
```

```python
# use numpy mean function
print(f'week 1 average= {np.mean(week_1)}')
# use numpy median function
print(f' week 1 median= {np.median(week_1)}')
print (f' week 1 mode = {mode(week_1)}')
```

```python
def central_tendency(temp):
   mean_temp = np.mean(temp)
   median_temp = np.median(temp)
   mode_temp = mode(temp)
   return mean_temp, median_temp, mode_temp
```

```python
mean_val, median_val, mode_val = central_tendency(week_1)

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
def central_tendency(temp,verbose=True):
   mean_temp = np.mean(temp)
   median_temp = np.median(temp)
   mode_temp = mode(temp)
   if verbose:
        print(f"The mean is {mean_temp}.")
        print(f"The median is {median_temp}.")
        print(f"The mode is {mode_temp}.")
   return mean_temp, median_temp, mode_temp
```

```python
print("\n=== Week 2 Test: verbose=True ===")
central_tendency(week_2, verbose=True)
```

```python
print("\n=== Week 2 Test: verbose=False ===")
mean_val, median_val, mode_val = central_tendency(week_2, verbose=False)
print("Returned values:", mean_val, median_val, mode_val)
```
